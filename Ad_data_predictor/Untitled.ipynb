```python
import numpy as np
import pandas as pd
import mglearn
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("adult.data",header=None, index_col=False,names=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','gender','capital-gain','capital-loss','hours-per-week','native-country','income'] )
data.head()
```

```python
print(data.gender.value_counts())
```

```python
features=['age','workclass','education','gender','hours-per-week','occupation']
x=pd.get_dummies(data[features])
yd=pd.get_dummies(data['income'])
x.head()
```

```python
y=yd[' >50K'].values
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kc=KNeighborsClassifier(n_neighbors=3)
kc.fit(x_train,y_train)
print("Training score:\n{:.2f}".format(kc.score(x_train,y_train)))
print("Testing score:\n{:.2f}".format(kc.score(x_test,y_test)))
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression(C=10000)
logreg.fit(x_train,y_train)
```

```python
print("Training score:\n{:.2f}".format(logreg.score(x_train,y_train)))
print("Testing score:\n{:.2f}".format(logreg.score(x_test,y_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
# print("{}".format(list(x_train.columns)))
clf=RandomForestClassifier(n_estimators=1000,random_state=20,max_depth=7,max_features=8)
clf.fit(x_train,y_train)
print("Training score:\n{:.2f}".format(clf.score(x_train,y_train)))
print("Testing score:\n{:.2f}".format(clf.score(x_test,y_test)))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbrt=GradientBoostingClassifier(random_state=23)
gbrt.fit(x_train,y_train)
print("Training score:\n{:.2f}".format(gbrt.score(x_train,y_train)))
print("Testing score:\n{:.2f}".format(gbrt.score(x_test,y_test)))
```
